```python
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error = [x + y for x,y in zip(variance, bias_squared)]

xs = range(len(variance))

plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias^2')
plt.plot(xs, total_error, 'b:', label='total error')

plt.legend(loc=9)
plt.xlabel("model complexity")
plt.title("The Bias-Variace Tradeoff")
plt.show()
```

```python
from __future__ import division
```

```python
def mean(x):
	return sum(x) / len(x)
```

```python
def median(v):
	n = len(v)
	sorted_v = sorted(v)
	midpoint = n//2
	if n % 2 == 1:
		return sorted_v[midpoint]
	else:
		lo = midpoint - 1
		hi = midpoint
		return (sorted_v[lo] + sorted_v[hi]) / 2
```

```python
def quantile(x, p): # 분위수
	p_index = int(p * len(x))
	return sorted(x)[p_index]
```

```python
def mode(x): # 최빈값
	counts = Counter(x)
	max_count = max(counts.values())
	return [x_i for x_i, count in counts.items() if count == max_count]
```

```python
def date_range(x):
	return max(x) - min(x)
```

```python
def sum_of_squares(vec):
	return sum([elem**2 for elem in vec])
def de_mean(x):
	x_bar = mean(x)
	return [x_i - x_bar for x_i in x]
def variance(x): # 분산
	n = len(x)
	deviations = de_mean(x)
	return sum_of_squares(deviations) / (n - 1)
```

```python
def standard_deviation(x): # 표준편차
	return math.sqrt(variance(x))
```

```python
def interquartile_range(x): # 사분위수 범위
	return quantile(x, 0.75) - quantile(x, 0.25)
```

```python
def covariance(x, y): # 공분산
	n = len(x)
	return np.dot(de_mean(x), de_mean(y) / (n - 1))
```

```python
def correlation(x, y): # 상관계수 이상치 제거 전
	stdev_x = standard_deviation(x)
	stdev_y = standard_deviation(y)
	if stdev_x > 0 and stdev_y > 0 :
		return covariance(x, y) / stdev_x / stdev_y
	else:
		return 0
```

```python
def normal_pdf(x, mu=0, sigma=1): # 확률밀도함수
	sqrt_two_pi = math.sqrt(2 * math.pi)
	return (math.exp(-(x-mu) ** 2 / 2 / sigma ** 2) / (sqrt_two_pi * sigma))
```

```python
def normal_cdf(x, mu=0, sigma=1): # 밀도함수
	return (1 + math.erf((x - mu) / math.sqrt(2) / sigma)) / 2
```

```python
def inverse_normal_cdf(p, mu=0, sigma=1, tolerance=0.00001): # 분위수함수
	if mu != 0 or sigma != 1:
		return mu + sigma * inverse_normal_cdf(p, tolerance=tolerance)
	low_z, low_p = -10.0, 0
	hi_z, hi_p = 10.0, 1
	while hi_z - low_z > tolerance:
		mid_z = (low_z + hi_z) / 2
		mid_p = normal_cdf(mid_z)
		if mid_p < p:
			low_z, low_p = mid_z, mid_p
		elif mid_p > p:
			hi_z, hi_p = mid_z, mid_p
		else:
			break
	return mid_z
```
